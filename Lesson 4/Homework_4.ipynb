```python
# # 1. Разобраться как использовать мультииндексные ключи в данном примере
index = [
    ('city_1', 2010),

    ('city_1', 2020),

    ('city_2', 2010),

    ('city_2', 2010),

    ('city_3', 2010),

    ('city_3', 2020),
    ]

population  = [
    101,
    201,
    101,
    201,
    101,
    201,
]
pop = pd.Series(population, index = index)
print(pop)

index = pd.MultiIndex.from_tuples(index)
pop = pop.reindex(index)
print(pop)


pop_df = pd.DataFrame(
    {
    'total': pop,
    'something':[
        10,
        11,
        12,
        13,
        14,
        15,
    ]
}
)
print(pop_df)
print()
print(pop_df['something'])
print()
pop_df_1 = pop_df.loc['city_1', 'something']

pop_df_2 = pop_df.loc[['city_1', 'city_3'], 'something']

pop_df_3 = pop_df.loc[['city_1', 'city_3'], ['something', 'total']]

print(pop_df_1)
print()
print(pop_df_2)
print()
print(pop_df_3)
```

```python
# 2. Из получившихся данных выбрать данные по
# - 2020 году (для всех столбцов)
# - job_1 (для всех строк)
# - для city_1 и job_2


index = pd.MultiIndex.from_product(
    [
        ['city_1', 'city_2'],
        [2010, 2020]
    ],
    names = ['city', 'year']
)
print(index)

columns = pd.MultiIndex.from_product(
    [
        ['person_1', 'person_2', 'person_3'],
        ['job_1', 'job_2']
    ],
    names = ['worker', 'job']
)

rng = np.random.default_rng(1)
data = rng.random((4,6))
print(data)

data_df = pd.DataFrame(data, index = index, columns = columns)
print(data_df)
print()
print('2. Из получившихся данных выбрать данные по ')
print()
print('- 2020 году (для всех столбцов)')
data_df_1 = data_df.loc[(slice(None), 2020), :]
print(data_df_1)
print()
print('- job_1 (для всех строк)')
data_df_2 = data_df.loc[:, (slice(None), 'job_1')]
print(data_df_2)
print()
print('- для city_1 и job_2 ')
data_df_3 = data_df.loc['city_1', (slice(None), 'job_2')]
print(data_df_3)
```

```python
# 3. Взять за основу DataFrame со следующей структурой
index = pd.MultiIndex.from_product(
    [
        ['city_1', 'city_2'],
        [2010, 2020]
    ],
    names=['city', 'year']
)
columns = pd.MultiIndex.from_product(
    [
        ['person_1', 'person_2', 'person_3'],
        ['job_1', 'job_2']
    ],
    names=['worker', 'job']
)

# Выполнить запрос на получение следующих данных
# - все данные по person_1 и person_3
# - все данные по первому городу и первым двум person-ам (с использование срезов)


rng = np.random.default_rng(1)
data = rng.random((4,6))
print(data)

data_df = pd.DataFrame(data, index = index, columns = columns)
print(data_df)

print('3. Выполнить запрос на получение следующих данных ')
print()
print('- все данные по person_1 и person_3')
data_df_1 = data_df.loc[:, (['person_1','person_3'], slice(None))]
print(data_df_1)
print()

print('все данные по первому городу и первым двум person-ам (с использование срезов)')
data_df_2 = data_df.loc[('city_1', slice(None)), (['person_1', 'person_2'], slice(None))]
print(data_df_2)
print()


# Приведите пример (самостоятельно) с использованием pd.IndexSlice
idx = pd.IndexSlice

print("Приведите пример (самостоятельно) с использованием pd.IndexSlice\n")

print("- Все данные по person_1 и person_3")
data_df_3 = data_df.loc[:, idx[['person_1', 'person_3'], :]]
print(data_df_3)
print()

print("- Все данные по первому городу и первым двум person-ам")
data_df_4 = data_df.loc[
    idx['city_1', :],
    idx[:'person_2', :]
]
print(data_df_4)
```

```python
#4. Привести пример использования inner и outer джойнов для Series (данные примера скорее всего нужно изменить)
ser1 = pd.Series(['a', 'b', 'c'], index=[1,2,3])
ser2 = pd.Series(['b', 'c', 'f'], index=[2,3,4])

print("Outer join (объединение всех):")
print(pd.concat([ser1, ser2], axis=1, join='outer'))

print("Inner join (только общие):")
print(pd.concat([ser1, ser2], axis=1, join='inner'))


# print (pd.concat([ser1, ser2], join='outer'))
# print (pd.concat([ser1, ser2], join='inner'))
```
